```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("clean_abalone.csv")
data.head()
```

```python
a = sns.FacetGrid(data, hue = 'Age', aspect=4 )
a.map(sns.kdeplot, 'Length', shade= True )
a.set(xlim=(0 , data['Length'].max()))
a.add_legend()
```

```python
a = sns.FacetGrid(data, hue = 'Age', aspect=4 )
a.map(sns.kdeplot, 'Diameter', shade= True )
a.set(xlim=(0 , data['Diameter'].max()))
a.add_legend()
```

```python
a = sns.FacetGrid(data, hue = 'Age', aspect=4 )
a.map(sns.kdeplot, 'Height', shade= True )
a.set(xlim=(0 , 0.25))#data['Height'].max()))
a.add_legend()
#ax = sns.kdeplot(setosa.sepal_width, setosa.sepal_length,cmap="Reds", shade=True, shade_lowest=False)
```

```python
g = sns.FacetGrid(data, col='Sex', hue="Age", height = 5, aspect=1)
g.map(sns.kdeplot, 'Height', shade=True)
g.set(xlim=(0 , 0.25))
g.add_legend();
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(data.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```
