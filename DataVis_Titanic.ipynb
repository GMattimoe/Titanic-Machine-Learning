```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("clean_titanic_WithNewAttributes.csv")
data.head()
```

```python
f,ax = plt.subplots(1,3,figsize=(18,6))
sns.countplot('Pclass',hue='Survived',data=data,palette='husl', ax=ax[1])
sns.countplot('Sex',hue='Survived',data=data,palette='husl', ax=ax[0])
sns.countplot('Title',hue='Survived',data=data,palette='husl', ax=ax[2])
ax[0].set_title('Survival by Gender')
ax[1].set_title('Survival by Class')
ax[2].set_title('Survival by Title')
```

```python
f,ax = plt.subplots(1,3,figsize=(18,6))
sns.countplot('SibSp',hue='Survived',data=data,palette='husl', ax=ax[0])
sns.countplot('Parch',hue='Survived',data=data,palette='husl', ax=ax[1])
sns.countplot('FamilySize',hue='Survived',data=data,palette='husl', ax=ax[2])

ax[0].set_title('Survival by Num Siblings and Spouses on Board')
ax[1].set_title('Survival by Num of Parents and Children on Board')
ax[2].set_title('Survival by Family Size on Board')
```

```python
a = sns.FacetGrid(data, hue = 'Survived', aspect=4 )
a.map(sns.kdeplot, 'Age', shade= True )
a.set(xlim=(0 , data['Age'].max()))
a.add_legend()
```

```python
a = sns.FacetGrid(data, hue = 'Survived', aspect=4 )
a.map(sns.kdeplot, 'Fare', shade= True )
a.set(xlim=(0 , data['Age'].max()))
a.add_legend()
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(data.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```
